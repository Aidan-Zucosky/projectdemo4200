```python
import seaborn as sns
import matplotlib.pyplot as plt

titanic_data = sns.load_dataset('titanic')
print(titanic_data.head())
```

```python
# Make three visualizations on the data set. At least 1 needs to be customized!
```

```python
town_count = titanic_data['embark_town'].value_counts(normalize = True)
plt.pie(town_count, labels = town_count.index, colors = ['blue', 'red', 'pink'], autopct='%1.1f%%')
plt.title("Where People on the Titanic Were From")
```

```python
A_titanic_data = titanic_data[titanic_data['alive'] == 'yes']
Awho_count = A_titanic_data['who'].value_counts() 

sns.barplot(x=Awho_count.index, y=Awho_count.values)
plt.title("Demogrphics of Survivors of the Titanic")
plt.ylabel("Number of People")
plt.show()
```

```python
A_titanic_data = titanic_data[titanic_data['alive'] == 'yes']
Adeck_count = A_titanic_data['deck'].value_counts() 

sns.barplot(x=Adeck_count.index, y=Adeck_count.values)
plt.title("What Deck of the Titanic Were Survivors On?")
plt.ylabel("Number of People")
plt.xlabel("Deck")
plt.show()
```
